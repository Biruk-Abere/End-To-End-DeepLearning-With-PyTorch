## Operations

The common standard arithmetic operators (+ , - , * , / and **) have all be lifted to element wise operations for any identically-shaped tensors of arbitrary shape. We can call element wise operations on any two tensors of the same shape. 

````python
import torch
x = torch.tensor([1.0 , 2 , 4 , 8])
y = torch.tensor([2 , 2 , 2 , 2])
print("The element wise sum value are :" , x+y)
print('The element wise subtraction are :' , x-y)
print("The element wise multiplication are :" , x*y)
print("The element wise division are :" , x/y)
print('The element wise power are :' , x**y)
````

## Concatenating Multiple Tensors 

We can also concatenate multiple tensors together , stacking them end to end to form a larger tensor. We just need to provide a list of tensors and tell the system along which axis to concatenate . The example below shows what happens when we concatenate two matrices along colums (axis 0 ) vs rows (axis = 1). 

````python
import torch
tensor1 = torch.arange(12 , dtype = torch.float32).reshape(3 , 4)
tensor2 = torch.tensor([[1 , 2 , 3 , 4 ] , [5 , 6 , 7 , 8] , [4 , 3 , 2 , 5]])
print("The shape of the tensor1 is :" , tensor1.shape)
print("The shape of the tensor2 is :" , tensor2.shape)
horizontal_concat = torch.cat((tensor1 , tensor2) , dim = 0)
vertical_concat = torch.cat((tensor1 , tensor2) , dim = 1)
print(vertical_concat.shape)
print(horizontal_concat.shape)
print("The value of the horizontal concat is \n :" , horizontal_concat)
print("The value of the vertical concat is \n :" , vertical_concat)
````

We saw how to perform element wise operations on two tensors of the same shape. Under certain condition , even when shapes differ , we can still perform element wise operations by invoking the broadcasting mechanism. This mechanism works in the following way. First expand one or both arrays by copying elements appropriately so that this transformations , the two tensors have the same shape , second carry out the element wise operations on the resulting arrays. 

````python
import torch
x = torch.arange(3).reshape(3 , 1)
y = torch.arange(2).reshape(1 , 2)
print("The shape of the tensor x is :" , x.shape)
print("The shape of the tensor y is :" , y.shape)
# now we can perform an element wise operation 
z = x + y
print("We can broadcast both of the tensors :" , z.shape)
````

## Saving Memory 

Running operations can cause new memory to be allocated to host results , for example , if we write y =  x + y , we will de-reference the tensor that Y used to point to and instead point Y at the newly allocated memory. In the following example , we demonstrate this with python id() function , which gives us the exact address of the referenced object in memory. After running y = y - x we will find that id(y) points to a different location. That is because python first evaluates Y+X , allocating a new memory for the result and then makes y point to this new location in memory. 

````python
import torch
x = torch.tensor([1 , 2 , 3 , 4])
y = torch.tensor([4 , 5 , 6 , 7])
before = id(y)
print("The id for the tensor x is :" , id(x))
print("The id for the tensor y is :" , id(y))
y = x + y
after = id(y)
print("The id after the operation is :" , id(y))
````

````python
before == after 
````

This might be undesirable for two reasons , first we do not want to run around allocating memory unnecessarily all the time. In machine learning we might have hundreds of megabytes of parameters and update all of them multiple times per second. Typically , we will want to perform these updates in place. Fortunately performing in-place operations is easy. we can assign the result of an operation to a previously allocated array with slice notation Example:- Y[:] = <expression>. To illustrate this concept we first create a new matrix Z with the shape as another Y , using zeros_like to allocate a block of 0 entries. 

````python
import torch
x = torch.tensor([1 , 2 , 3 , 4])
y = torch.tensor([4 , 5 , 6 , 7])
# =================================
z = torch.zeros_like(y)
before = id(z)
z[:] = x + y
after = id(z)
before == after
````

# Moving To GPU 

One of the major advantages of PyTorch is its robust acceleration on CUDA compatible NVDIA GPU's(CUDA stands for compute unified device architecture) which is NVIDIA's platform for parallel computing ? 

So far everything we've done has been on CPU. How do we move to the faster hardware ? First we should check whether a GPU is available , with the is_available() method. 

Note:- If we do not have a CUDA compatible GPU and CUDA drivers installed , the executable cells in this section will not execute any GPU related code. 

````python
import torch
if torch.cuda.is_available():
  print("cuda is available")
else:
  print("Cuda is not available")
````

Once we've determined that one or more GPU's available. We need to put our data someplace where the GPU can see it. Our cpu does computation in our computer's RAM , our GPU has dedicated memory detached to it. Whenever we want to perform a computation on a device , we must move all the data needed for that computation to memory accessible by that device.  

There are multiple ways to get our data on to our target device. we may do it a creation time. 

````python
import torch
if torch.cuda.is_available():
  random_cuda = torch.rand((2,2) , device = "cuda")
  print(random_cuda)
else:
  print("Only CPU is available")
````

By default, new tensors are created on the CPU , so we have to specify when we want to create our tensor on the GPU with the optimal device argument. we can see when we print the new tensor , PyTorch informs us which device it's on (if it is not 
on the CPU)

* We can query the number of the GPU's with torch.cuda.device_count() , if we have more than one GPU , we can specify them by index , device = "cuda:0" , device = "cuda:1" etc...

* As a coding practice specifying our devices everywhere with string constraints is pretty fragile , in an ideal world , our code would perform robustly wheather we are on CPU or GPU hardware. we can do this by creating a device handle that can be passed to our tensors instead of string.    



````python
import torch 
if torch.cuda.is_available():
  device_count = torch.cuda.device_count()
  print("Your number of devices are = {}".format(device_count))
else:
  print("Cpu only used")
````

````python
import torch
if torch.cuda.is_available():
  my_device = torch.device("cuda")
else:
  my_device = torch.device("cpu")

tensor_device =  torch.rand((2 , 2) , device = my_device)
print(tensor_device)
print("My tensor device is :" , tensor_device.device)
````

If we have an existing tensor living on one device , we can move it to another with the to() method. The following line of code creates a tensor on the CPU and moves it to which ever device handle we acquired in the previous cell. 

````python
import torch
my_tensor = torch.tensor([1 , 2 , 3 , 4])
my_tensor.to(my_device)
````

It is important to know that inorder ro do computation involving two or more tensor , all of the tensors must be on the same device. The following code will throw a run time error regardless of wheather we have a GPU device available. 

````python
import torch
tensor_CPU = torch.tensor([1 , 2 , 3 , 4 ])
tensor_GPU = torch.tensor([1 , 2 , 3 , 4] , device = my_device)
sum = tensor_CPU + tensor_GPU
print("We can't operate tensors that are not on different devices = {}".format(sum)) # exception will be thrown
````

````python
# another example of using view in the PyTorch model
import torch
class Net(torch.nn.Module):
  def __init__(self):
    super().__init__()
    self.conv1 = torch.nn.Conv2d(in_channels = 3 , out_channels = 6 , kernel_size = (5,5))
    self.activation1 = torch.nn.ReLU()
    self.pool1 = torch.nn.MaxPool2d(kernel_size = (2 , 2) , stride = (2 ,2))
    self.conv2 =  torch.nn.Conv2d(in_channels=6 , out_channels = 16 , kernel_size = (5 ,5))
    self.activation2 = torch.nn.ReLU()
    self.pool2 = torch.nn.MaxPool2d(kernel_size = (2,2) , stride = (2 ,2))
    self.fc1 = torch.nn.Linear(in_features = 16*5*5 , out_features = 120)
    self.activation3 = torch.nn.ReLU()
    self.fc2 = torch.nn.Linear(in_features = 120  , out_features = 84)
    self.activation4 = torch.nn.ReLU()
    self.fc3 = torch.nn.Linear(in_features = 84 , out_features = 10)
    self.softmax = torch.nn.Softmax(dim = 0)

  def forward(self , x):
    x = self.pool1(self.activation1(self.conv1(x)))
    x = self.pool2(self.activation2(self.conv2(x)))
    x = x.view(-1 , 16*5*5) # PyTorch will automatically calcualates the dimension
    x = self.activation3(self.fc1(x))
    x = self.activation4(self.fc2(x))
    x = self.softmax(self.fc3(x))
    return x



# Creating the object
model = Net()
print("The entire layers are \n :" , model)
print("The shape of the entire parameters of the model \n " )
for params in model.parameters():
  print(params.shape)
````

## How to Build PyTorch Models Easily Using View

````python
import torch

class MNIST_MODEL(torch.nn.Module):
  def __init__(self , input_size , hidden_size , output_size):
    super().__init__()
    # hidden layer
    self.linear1 = torch.nn.Linear(in_features = input_size , out_features = hidden_size)
    # Activation function
    self.activation1 = torch.nn.ReLU()
    # output layer 
    self.linear2 = torch.nn.Linear(in_features = hidden_size , out_features = output_size)
    # Activation Function
    self.softmax= torch.nn.Softmax(dim=0)
  
  def forward(self , x):
    x = x.view(x.size(0) , -1)
    x = self.linear1(x)
    x = self.activation1(x)
    x = self.linear2(x)
    x = self.softmax(x)
    return x

# Creating an object
input_size = 784
num_classes = 10
model = MNIST_MODEL(input_size , hidden_size = 32 , output_size = num_classes)

# let's look the parameters
print("The shape of the parameters , the weights and biases for each layer ")
for params in model.parameters():
  print(params.shape)
````

````python
import torch
class LeNet(torch.nn.Module):
  def __init__(self , input_channels , number_classes):
    # call the parent constructor 
    super().__init__()
    # Initialize first set of conv -> ReLu-> Pool layers
    self.convd1 = torch.nn.Conv2d(in_channels = input_channels , out_channels = 20 , kernel_size = (5,5))
    self.relu1 =  torch.nn.ReLU()
    self.maxpool1 = torch.nn.MaxPool2d(kernel_size = (2 ,2) , stride = (2,2))

    # Initialize the second set of Conv -> RELU -> Pool layer
    self.convd2 = torch.nn.Conv2d(in_channels = 20 , out_channels = 50 , kernel_size = (5,5))
    self.relu2  = torch.nn.ReLU()
    self.maxpool2 = torch.nn.MaxPool2d(kernel_size = (2 , 2) , stride = (2 , 2 ))

    # Initialize first and set FC -> ReLu layers
    self.flatten = torch.nn.Flatten()
    self.fc1 = torch.nn.Linear(in_features = 800 , out_features = 500)
    self.activation = torch.nn.ReLU()

    # Initialize our softmax classifier
    self.Fc2 = torch.nn.Linear(in_features = 500 , out_features = number_classes)
    self.logsoftmax = torch.nn.LogSoftmax(dim=1)


    def forward(self , x):
      x = self.conv1(x)
      x = self.relu1(x)
      x = self.maxpool1(x)

      # pass the output from the previous layer through the second set of conv -> ReLu -> Pool layers
      x = self.conv2(x)
      x = self.relu2(x)
      x = self.maxpool2(x)

      # Flatten the output from the previous and pass it through our only set of FC -> ReLu layers
      x = self.flatten(x , 1)
      x = self.Fc1(x)
      x = self.relu3(x)
      x = self.Fc2(x)
      output = self.logsoftmax(x)
      return output

model = LeNet(input_channels = 1 , number_classes = 10)
print("The entire model ")
print(model)

print("Specificially a single layer")
print(model.convd1)

print("Printing the entire parameters")
for params in model.parameters():
  print(params.shape)

print("Printing a specific parameters")
for params in model.convd2.parameters():
  print(params.shape)
````

## Building Convolutional Neural Network using Sequential and torch.nn.Module base class

````python
import torch
class LeNet(torch.nn.Module):
  def __init__(self , input_channels , number_classes):
    super().__init__()
    self.convolution_stack = torch.nn.Sequential(
        torch.nn.Conv2d(in_channels = input_channels , out_channels = 20 , kernel_size = (5,5)),
        torch.nn.ReLU(),
        torch.nn.MaxPool2d(kernel_size = (2,2) , stride = (2,2)) ,
        
        
        torch.nn.Conv2d(in_channels = 20 , out_channels = 50 , kernel_size = (5,5)),
        torch.nn.ReLU(),
        torch.nn.MaxPool2d(kernel_size = (2,2) , stride = (2,2)),

        torch.nn.Flatten(),
        torch.nn.Linear(in_features = 500 , out_features = 800),
        torch.nn.ReLU(),
        torch.nn.Linear(in_features = 800 , out_features = number_classes),
        torch.nn.Softmax(dim = 1)
    )
  def forward(self , x):
    convs = self.convolution_stack(x)
    return convs


# Creating an object 
model = LeNet(input_channels = 1, number_classes = 10)
print('THE ENTIRE MODEL STRUCTURE')
print(model)

print("Print a single layer")
print(model.convolution_stack[1])

print("Print the shape of the entire model parameters")
for params in model.parameters():
  print(params.shape)

print("print the shape of a single layer model parameters")
for params in model.convolution_stack[1].parameters():
  print(params.shape)
````

# Mean Absolute Error (L1 Loss Function)

The L1 loss function computes the mean absolute error between each value in the predicted tensor and that of the target. It first calculates the absolute difference between each value in the predicted tensor and that of the target and computes the sum of all the values returned from each absolute difference computation. Finally it computes the average of this sum value to obtain the mean absolute error(MAE). The L1 loss function is very robust for handling noise.

# Mean Squared Error Loss (L2 Loss Function)

Measures the amount of error in statistical methods. It assesses the average squared difference between the observed and the predicted values. When a model has no error , the MSE equals to zero. As model error increases , its value increases. The mean squared error is also known as the mean squared deviation(MSD)

Forexample :- in regression , the mean squared error represent the average squared residual. 

image.png

image.png

The calculations for the mean squared error are similar to the variance. To find the MSE , take the observed value , subtract the predicted value ans square that difference , Repeat that for all the observations , then sum of all those values and divide by the number of observation. 

The MSE is the average squared distance between the observed and the predicted values. Squaring the difference serves several purposes :- 

Squaring the differences eliminates negative values for the differences which ensures that the mean squared error is always greater than or equal to zero. It is almost always a positive value. only a perfect model with no error produces an MSE of zero and it doesn't occur in practice. 

Additionally squaring increases the impact of large errors. These calculations disproportionally penalize larger errors more than smaller errors. This property is essential when we want our model to have smaller errors.

We define a device data loader clas to wrap our existing data loaders and move data to the selected device as batches are accessed. We shouldn't put all of our data at once on the GPU , our data might be a tens of GPU's in size. and we might not want to put all of them on to GPU , what we will do is just as we are about to pass some data in to a model at that very moment we should put it on to GPU. This way we will be sure that we don't busy the GPU when we don't need to. 

````python
# there is no base class to inherit in here
class DeviceDataLoader():
    def __init__(self , train_loader , device):
        self.train_loader =  train_loader 
        self.device = device
    def __iter__(self):
        """Moving the dataa to the selected device as batches are accessed"""
        for inputs , targets in self.train_loader:
            yield to_device(inputs , self.device)
            yield to_device(targets , self.device)
    def __len__(self):
        """Number Of Batches"""
        return len(self.train_loader)
   
# Creating an object
train_loader = DeviceDataLoader(train_loader = train_loader ,device = my_device)
test_loader =  DeviceDataLoader(train_loader = test_loader , device = my_device)
print("The train loader is on the device :" , train_loader.device)
print("The test loader is on the device : " , test_loader.device) 
````

While building real world machine learning models it is quite common to split the dataset in to 3 parts 

* ` Trainig set:- ` Used to train the model that is compute the loss and adjust the weights of the model using gradient descent

* `Validation set :- ` Used to evaluate the model while training , adjust hyperparameters (learning rate) and pick the best version of the model. 

* `Test set :-`Used to compare different models or different types of modeling approaches and report the final accuracy of the model. 

* In the MNIST dataset , there are 60,000 training images and 10,000 test images. The test set is standardized so that different researchers can report tje results of their models against the same set of images.
````python
import numpy as np
def split_indices(n , val_percent):
  # determine the size of the validation set
  n_val = int(val_percent * n)
  # create a random permutuation of 0 to n-1
  idxs = np.random.permutation(n)
  return idxs[n_val:] , idxs[:n_val]

train_indices , validation_indices = split_indices(len(data_train) , val_percent = 0.2)
print(len(train_indices) , len(validation_indices))
````

Split indices randomly shuffles the array indices 0 , 1 , ... , n-1 and separates out a desired portion from it for the validation set because the training images are often ordered by the target labels that is images of 0s , followed by images of 1s , followed by images of 2s and so on , if we were to pick 20% validation set simply by selecting the last 20% of the images , the validation set would only consist of images of 8s and 9s , where as the training set would not contains images of 8s and 9s , this would make it impossible to train a good model using the training set , which also performs well on the validation set.

# Training / Evaluating and Predicting 

Training :- As with logistic regression , we can use cross entropy as the loss function and accuracy as the evaluation metric for our model. The training loop is also identical to the previous , so we can resue the loss_batch , evaluate and fit functions. 

The loss batch function calculates the loss and metric value for a batch of data and optionally performs gradient if an optimizer is provided

````python
def loss_batch(model , loss_func , xb , yb , opt = None , metric = None):
  # Generate predictions 
  predictions = model(xb)
  # calculate the loss
  loss = loss_func(predictions , yb)
  if opt is not None:
    # compute gradients
    loss.backward()
    # update the parameters
    opt.step()
    #reset the gradients
    opt.zero_grad()

  metric_result = None
  if metric is not None:
    # compute the metric
    metric_result = metric(predictions , yb)
  
  return loss.item() , len(xb) , metric_result

````

The evaluate function calculates the overall loss and metric for the validation set

````python
def evaluate(model , loss_fn , valid_dl , metric = None):
  with torch.no_grad():
    # pass each batch through the model
    results = [loss_batch(model , loss_fn , xb , yb , metric = metric) for xb , yb in valid_dl]
    # separate the losses , counts and metrics
    losses , nums , metrics = zip(*results)
    # Total size of the dataset
    total = np.sum(nums)
    # Average loss across the batches
    average_loss = np.sum(np.multiply(losses , nums))/total

    avg_metric = None
    if metric is not None:
      # average of metric accross batches 
      avg_metric = np.sum(np.multiply(metrics , nums))/total

    return avg_loss , total , avg_metric
````

The fit function contains the actual training loop , Now 
* Instead of defining the optimizer manually , we'll pass in the learning rate and create an optimizer inside the fit function. This will allow us to train the model with different learning rates , if required

* We will record the validation loss and accuracy at the end of every epochs and return the history as output of the fit function. 

````python
def fit(epochs , lr , model , loss_fn , train_loader , validation_loader , metric = None , opt_fn = None):
  losses , metrics = [] , []
  if opt_fn is None: opt_fn = torch.optim.SGD
  opt = torch.optim.SGD(model.parameters() , lr = lr)
  for epoch in range(epochs):
    # Training 
    for xb , yb in train_loader:
      loss , _ , _ =  loss_batch(model , loss_fn , xb , yb , opt)
    # Evaluation
    result = evaluate(model , loss_fn , validation_loader , metric)
    val_loss , total , val_metric = result
    # Record the loss and metric
    losses.append(val_loss)
    metrics.append(val_metric)

    # print the progress
    if metric is None:
      print('Epoch [{}|{}] , Loss :{:.4f}' , format(epoch+1 , epochs , val_loss))
    else:
      print('Epoch[{}|{}] , loss :{:.4f} , {}:{:.4f}' , format(epoch +1 , epochs , val_loss , metric.__name__ , val_metric))
  return losses , metrics
````

we will also define an accuracy function which will calculate the overall accuracy of the model on an entire batch of outputs , so that we can use it as a metric in fit

````python
def accuracy(outputs , labels):
  _ , preds =  torch.max(outputs , dim = 1 )
  return torch.sum(preds == labels).item()/len(preds)
````

# PyTorch Model Evaluation 

* What is model evaluation ?
* Why do i need a model evaluation ?
* How to implement model evaluation in PyTorch ? 

Once the model is fit , it can be evaluated on the test dataset. This can be achieved by using the DataLoader for the test set , then comparing the predictions to the expected  values of the test set and calculating the performance metric. 

`Making Predctions ` :- A fit model can be used to make a prediction on a new data. Forexample :- we might have a single image or a single rows of data and want to make a prediction. 

There are certain layers that behave differently in training and evaluation models such as the dropout and batch normalize layer. 

`Dropout layer :- ` The drop out layer use for regularization , with the argument 'p' that denotes the drop probability , which is used to determine the probability of dropping the units during training. 

* One important point to remember is that units may drop randomly during training only , for the evaluation or inference phase , all the hidden units must be active. To ensure that the overall activations are on the same scale during training and predictions , the activations of the active neurons have to be scaled appropriately. 

* When calling this layer , its behaviour can be controlled via model.train() and model.eval() to specify whether this call will be made during training and during inference , when using drop out , alternating between these two modes is crucial to ensure that is behaves correctly , for instance , nodes are only randomly dropped during training not evaluation or inference.

`Batch Normalization Layer :- ` The PyTorch API provides a class , nn.BatchNorm2d() that we can use a layer when defining our models. Note that the behaviour of updating the learnable parameters depends on whether the model is a training model or not , these parameters are learned only during training are then used for noramlization during evaluation. because normalization used for evaluation(testing)

* Note that using the designated setting for training model.train() and evaluation model.eval() will automatically set the mode for the drop out layer and batch normalization layers and rescale appropriately so that we do not have to worry about that at all. 

Using `with torch.no_grad():-` The validation loop looks very similar to training but is somewhat simplified. The key difference is that validation is read only , specifically the loss value returned is not used(We don't use the loss functions for gradient purpose) and weights are not updated. 

````python
# Check the accuracy on the test set 
def check_accuracy(loader , model):
  num_correct = 0
  num_samples = 0
  model.eval()

  with torch.no_grad():
    for inputs , labels in loader:
      inputs = to_device(inputs , my_device)
      labels = to_device(labels , my_device)
      scores = model(inputs)
      _ , predictions = scores.max(1)
      num_correct += (predictions == labels).sum().item()
      num_samples += labels.size(0)
  
  print("The accuracy of this model on the test is ")
  acc = (num_correct / num_samples)*100
  return acc

test_accuracy = check_accuracy(test_loader , model = model)
print("our test accuracy is :" , test_accuracy)
````
